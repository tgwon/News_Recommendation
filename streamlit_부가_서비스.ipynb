```python
#필요하다면 설치 후 import
import pandas as pd
import random
import matplotlib.pyplot as plt


# path를 자신의 경로에 맞게 지정!
path = 'C:/Users/LG/Desktop/대구 빅데이터/streamlit/data/'
```

```python
# CSV 파일을 데이터프레임으로 불러오기
# parse_list 함수를 포함하여 실행시켜줘야합니다.


def parse_list(input_str):
    return eval(input_str)

client_fv= pd.read_csv(path + "client_feature_vector.csv", converters={'feature': parse_list})
daily_result = pd.read_csv(path + "daily_result.csv", converters={'fv': parse_list})
cw = pd.read_csv(path + "consumer_word.csv")
```

```python
#오늘의 소비자 단어

num_indices = len(cw)

# 랜덤으로 인덱스 3개 추출
random_indices = random.sample(range(num_indices), 3)

print(f'"{cw.단어[random_indices[0]]}"')
print(f'의미 : {cw.뜻[random_indices[0]]}')
```

```python
#오늘의 뉴스 현황

value_counts = daily_result['category'].value_counts()

fig2=plt.figure(figsize=(10,5))

#한글 폰트 설정
plt.rcParams['font.family'] = 'HYdnkM'


value_counts.plot(kind='bar', color='skyblue')
plt.title('오늘의 뉴스 현황', fontsize=25)
plt.rc('xtick', labelsize=14) 
plt.xlabel('')
plt.xticks(rotation=0)
for i, j in zip([0,1,2,3,4,5,6,7], value_counts.values):
    plt.text(i, j, f'{int(j/len(daily_result)*100)}%', ha='center', va='bottom')
```
