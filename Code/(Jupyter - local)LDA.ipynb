```python
#필요하다면 설치 후 import
import pandas as pd
import itertools
from collections import Counter
from gensim import corpora
import gensim
import pyLDAvis
import pyLDAvis.gensim

# path를 자신의 경로에 맞게 지정!
path = 'C:/Users/LG/Desktop/대구 빅데이터/streamlit/data/'
```

```python
# CSV 파일을 데이터프레임으로 불러오기
# parse_list 함수를 포함하여 실행시켜줘야합니다.


def parse_list(input_str):
    return eval(input_str)

client_fv= pd.read_csv(path + "client_feature_vector.csv", converters={'feature': parse_list})
daily_result = pd.read_csv(path + "daily_result.csv", converters={'fv': parse_list})
```

```python
daily_result.head(1)
```

```python
client_fv.head(1)
```

```python
#LDA의 입력값으로 활용하기 위해 전처리된 텍스트를 띄어쓰기를 기준으로 토큰화
tokenized_doc = daily_result['content_lda'].apply(lambda x: x.split())
```

```python
# daily news들에 등장한 모든 단어들
k=(list(itertools.chain(*tokenized_doc)))

# daily news들에 등장한 단어 빈도수 세기, 상위 5개 단어 출력
vocab = Counter(k).most_common(5)
print("상위 5개 단어 :", vocab)
```

```python
# 토픽 모델링 딕셔너리 생성
# corpora.Dictionary 함수로 정수인코딩과 단어의 빈도수를 기록
dictionary= corpora.Dictionary(tokenized_doc)
```

```python
# (단어,빈도)
corpus = [dictionary.doc2bow(text) for text in tokenized_doc]
```

```python
print(corpus[1])

#ex첫번째 뉴스에서 39번 단어는 6번 등장, 45번 단어는 7번 등장
```

```python
#num_topis : 토픽 수 5개로 지정
#passes : 알고리즘의 동작 횟수,
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = 5,
                                           id2word=dictionary,
                                           passes=15)
```
